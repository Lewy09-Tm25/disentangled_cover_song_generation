```python
import os
from collections import Counter

def analyze_pairs(path):
    files = [f for f in os.listdir(path)]
    ids = []
    for f in files:
        if "_" in f:
            work_id = f.split("_")[0]
            ids.append(work_id)
    counts = Counter(ids)
    non_paired = {}
    for work_id, count in counts.items():
        if count != 2:
            non_paired[work_id] = count
    return non_paired

nps = analyze_pairs("chorus_audio")
# print(analyze_pairs("audio_downloads"))
```

```python
import os
import shutil
from collections import defaultdict

def move_songs(fpath, tpath):
    if not os.path.exists(tpath):
        os.makedirs(tpath)
        print("created target folder")

    groups = defaultdict(list)
    files = [f for f in os.listdir(fpath)]
    for f in files:
        if "_" in f:
            wid = f.split("_")[0]
            groups[wid].append(f)
    moved_count = 0

    for wid, fnames in groups.items():
        nfiles = len(fnames)
        if nfiles == 1:
            filetomove = fnames[0]
            shutil.move(os.path.join(fpath, filetomove), os.path.join(tpath, filetomove))
            moved_count += 1
        elif nfiles > 2:
            extras = fnames[2:]
            for extra in extras:
                shutil.move(os.path.join(fpath, extra), os.path.join(tpath, extra))
                moved_count += 1
    print("process done")

move_songs("audio_downloads", "extra_audio")
```
